## Haberman Exploratory Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('haberman.csv')

```

```python
# How many Data points and features ?
data.shape
```

```python
# what are the columns in our dataset ?
data.columns
```

```python
# how many data points for each class are present ?
# Balanced vs Unbalanced datasets
# since we can see 225 points belong to class 1 and 81 points belong to class 2 So it is a unbalanced data set.
data['survival_status'].value_counts()
```

```python
# cannot make sense out of it
data.plot(kind='scatter',x='age',y='axillary_node')
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(data, hue="survival_status", size=4) \
   .map(plt.scatter, "age", "axillary_node") \
   .add_legend();
plt.show();
```

```python
#pair_wise scatter plot:
# can be used when number of features are high
# plot 2-D scatter
# following are pairwise scatter on haberman datasets

plt.close();
sns.set_style("whitegrid");
sns.pairplot(data, hue="survival_status", size=3);
plt.show()
```

```python
sns.FacetGrid(data,hue='survival_status',size=5)\
    .map(sns.distplot,'age') \
    .add_legend();
plt.show()

```

```python
sns.FacetGrid(data,hue='survival_status',height=5) \
    .map(sns.distplot,'axillary_node') \
    .add_legend();
plt.plot()
```

```python
sns.FacetGrid(data,hue='survival_status',height=5) \
   .map(sns.distplot,'op_year') \
   .add_legend();
plt.plot()
```

```python
counts , bin_edges= np.histogram(data['age'],bins=10,density=True)
pdf=counts/sum(counts)

cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
sns.boxplot(x='survival_status',y='age',data=data)
plt.plot()
```

```python
sns.violinplot(x='survival_status',y='age',data=data)
plt.plot()
```
